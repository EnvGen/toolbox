```python
summary_df = None
for sample_set in ['lmo', 'baltic_transect_2014', 'baltic_redox_cline_2014']:
    print('Starting with ' + sample_set, flush=True)
    df = pd.read_table("../tables/{}/all_genes.raw_counts.tsv.gz".format(sample_set))
    print('Finished reading in ' + sample_set, flush=True)
    val_cols = list(df.columns)
    val_cols.remove('gene_id')
    val_cols.remove('gene_length')
    tmp_summary_df = df[val_cols].tail().T
    tmp_summary_df.columns = df['gene_id'].tail(5).values
    tmp_summary_df['total_sum'] = df[val_cols].sum()
    tmp_summary_df['with_feature'] = tmp_summary_df['total_sum'] - tmp_summary_df['__no_feature'] - tmp_summary_df['__ambiguous'] - tmp_summary_df['__not_aligned']
    if summary_df is None:
        print("ITS NONE!")
        summary_df = tmp_summary_df
    else:
        summary_df = summary_df.append(tmp_summary_df)
```

```python
unwanted_headers = ['Id', 'Name', 'Categories', 'Description', 'gene_length', 'ncbi-cdd', 'cazy-class', 'cazy-note', 'cazy-activities']
annotation_sources = ['EggNOG', 'pfam_uniq', 'EC', 'dbCAN', 'TIGRFAM']
```

```python
for annotation_source in annotation_sources:
    print('Starting with ' + annotation_source, flush=True)
    annotation_s = None
    for sample_set in ['lmo', 'baltic_transect_2014', 'baltic_redox_cline_2014']:
        df = pd.read_table("../tables/{}/{}.raw_counts.annotated.tsv.gz".format(sample_set, annotation_source), nrows=1000)
        val_cols = list(df.columns)
        for header in unwanted_headers:
            if header in val_cols:
                val_cols.remove(header)
        if annotation_s is None:
            annotation_s = df[val_cols].sum()
        else:
            annotation_s = annotation_s.append(df[val_cols].sum())
    summary_df[annotation_source] = annotation_s
```

```python
summary_df
```

```python
melt_plot_df = pd.melt(summary_df[annotation_sources].divide(summary_df['with_feature'],axis=0))
```

```python
melt_plot_df['Normalization'] = "Fraction of reads mapping to included genes"
```

```python
melt_plot_df2 = pd.melt(summary_df[annotation_sources].divide(summary_df['total_sum'],axis=0))
```

```python
melt_plot_df2['Normalization'] = "Fraction of all reads"
```

```python
melt_plot_df = melt_plot_df.append(melt_plot_df2)
```

```python
ax = seaborn.boxplot(data=melt_plot_df, y='value', x='variable', orient='v', hue='Normalization')
ax.set_xlabel('Annotation Source')
ax.set_ylabel('Fraction of Reads')
ax.set_xticklabels(['EggNOG', 'Pfam', 'EC', 'dbCAN', 'TIGRFAM'])
ax.set_ylim(0,0.5)
plt.savefig("boxplot_fraction_of_reads_per_annotation_source.pdf")
```
